```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
#loading the data
games = pd.read_csv('games.csv')
# games.round(6)
# games.replace([np.inf,-np.inf],np.nan)
```

```python
print(games.columns)
print(games.shape)
```

```python
plt.hist(games['average_rating'])
plt.show()
```

```python
print(games[games['average_rating']==0].iloc[0])
print(games[games['average_rating']>0].iloc[0])
```

```python
#Since there are nearly 25k games where the users did not rate the games, let's remove those from our data set
games = games[games["users_rated"]>0]
games.dropna(axis=0,inplace=True)# removing rows with missing values
plt.hist(games["average_rating"])
plt.show()
```

```python
corrmatrix = games.corr()
fig = plt.figure(figsize=(10,10))
sns.heatmap(corrmatrix, vmax=0.8, square = True)
plt.show()
```

```python
columns = games.columns.tolist()#get a list of columns in our data frame
columns = [c for c in columns if c not in ['bayes_average_rating','aveage_rating','id','name','type']]# remove columns that are not required for training
target = 'average_rating'#item to be predicted
```

```python
train = games.sample(frac=0.8,random_state = 1)
test = games.loc[~games.index.isin(train.index)]
print(train.shape)
print(test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
LR = LinearRegression()
LR.fit(train[columns], train[target])
```

```python
predictions = LR.predict(test[columns])
mean_squared_error(predictions,test[target])
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(n_estimators = 100, min_samples_leaf = 10, random_state = 1)
RFR.fit(train[columns],train[target])
```

```python
predictions = RFR.predict(test[columns])
mean_squared_error(predictions,test[target])
```

```python
test[columns].iloc[0]
```

```python
#Predict using the two models
rating_LR = LR.predict(test[columns].iloc[0].values.reshape(1,-1))
rating_RFR = RFR.predict(test[columns].iloc[0].values.reshape(1,-1))
print(rating_LR)
print(rating_RFR)
print("the actual rating: ",test[target].iloc[0])

rating_LR = LR.predict(test[columns].iloc[10].values.reshape(1,-1))
rating_RFR = RFR.predict(test[columns].iloc[10].values.reshape(1,-1))
print(rating_LR)
print(rating_RFR)
print("the actual rating: ",test[target].iloc[10])
```
